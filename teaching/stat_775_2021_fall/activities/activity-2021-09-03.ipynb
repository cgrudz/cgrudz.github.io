### Exercise 2:

The real strength of `matplotlib` is not in how we can combine lines in a single plot, but how flexible its object-oriented design is for changing every attribute of a figure.  We gain this functionality principally by using the [`fig` class](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html) and [`axis` class](https://matplotlib.org/stable/api/axes_api.html) API.  An example of creating a `fig` is below, in which we add `axis` attributes to the `fig` and henceforth set `axis` attributes to adjust the plot.

```python
import numpy
import matplotlib.pyplot

fig = plt.figure(figsize=(10.0, 3.0))

ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

ax1.set_ylabel('average')
ax1.plot(numpy.mean(data, axis=0))

ax2.set_ylabel('max')
ax2.plot(numpy.max(data, axis=0))

ax3.set_ylabel('min')
ax3.plot(np.min(data, axis=0))

fig.tight_layout()

plt.savefig('inflammation.png')
plt.show()
```
